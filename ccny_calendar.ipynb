```python
url = "https://www.ccny.cuny.edu/registrar/fall"
r = requests.get(url)
html_elemenets = r.text
print(f"Returns 200 if get request was successful: {r.status_code}")
#can parse
soup = BeautifulSoup(html_elemenets)
cal_title = soup.title
# print(cal_title)
cal_title = soup.title.text
print(f"Scraping calender from: {cal_title}")


#Alternativee
# df_as_list = pd.read_html(url)
# df = df_as_list[0]
# print(df)
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://www.ccny.cuny.edu/registrar/fall"
```

```python
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

table = soup.find("table")
rows = table.find_all("tr")
```

```python
data = []

for row in rows[1:]:  
    cells = row.find_all(["th", "td"])
    if len(cells) < 3:
        continue

    date_text = cells[0].get_text(" ", strip=True).replace("\xa0", " ")
    dow = cells[1].get_text(" ", strip=True)
    text = cells[2].get_text(" ", strip=True)

    first_day = date_text.split("-")[0].strip()

    if "," in first_day:
        date_value = pd.to_datetime(first_day).date()
    else:
        date_value = pd.to_datetime(first_day + " 2021").date()

    data.append({"date": date_value, "dow": dow, "text": text})
```

```python
calendar_df = pd.DataFrame(data)
calendar_df = calendar_df.sort_values("date").reset_index(drop=True)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.cuny.edu/academics/academic-calendars/fall-2021-academic-calendar/"
response = requests.get(url)
html = response.text
```

```python
soup = BeautifulSoup(html, "html.parser")
table = soup.find("table")
rows = table.find_all("tr")

dates = []
dow = []
texts = []
for row in rows[1:]:   
    columns = row.find_all("td")
       if len(columns) >= 3: 
        date = columns[0].get_text(strip=True)
        day  = columns[1].get_text(strip=True)
        text = columns[2].get_text(strip=True)
        
        if "-" in date:
            date = date.split("-")[0].strip()
        date = date + " 2021"
        dates.append(date)
        dow.append(day)
        texts.append(text)
```

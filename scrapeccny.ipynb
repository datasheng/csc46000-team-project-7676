```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.ccny.cuny.edu/registrar/fall"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
table = soup.find("table")
rows = table.find_all("tr")
print("Number of rows:", len(rows))
```

```python
data = []
for row in rows[1:]:
    cols = row.find_all("td")
    if len(cols) >= 3:
        date_str = cols[0].get_text(strip=True)
        dow_str  = cols[1].get_text(strip=True)
        text_str = cols[2].get_text(strip=True)
        data.append([date_str, dow_str, text_str])
print(data[:3])
```

```python
df = pd.DataFrame(data, columns=["date", "dow", "text"])
df.head()
```

```python
clean_dates = []
for d in df["date"]:
    d = d.replace(",", "")
    d = d.split("-")[0].strip()

    if ("2021" not in d) and ("2022" not in d):
        d = d + " 2021"
    clean_dates.append(pd.to_datetime(d))

df["date"] = clean_dates
df.head()
```

```python
df = df.set_index("date")
df = df.sort_index()
print(df.to_string(max_colwidth=100))
```
